## Task:- Identify if there are any restaurant chains  present in the dataset.

```python
restaurant_chains=df['Restaurant Name'].value_counts().reset_index(name="Branches")
#  new=restaurant_chains[restaurant_chains["Branches"]>1]
restaurant_chains.head(10)
```

```python
restaurant_chains.head(10).plot(x="Restaurant Name",y="Branches",kind='bar', color='blue')
plt.title('Top 10 Restaurant Chains by Branches')
plt.ylabel("Branches")
plt.show()
```

```python
ratings=df.groupby("Restaurant Name")["Aggregate rating"].mean().reset_index(name="Avg Rating") 
ratings=ratings.sort_values(by="Avg Rating",ascending=False)
ratings
```

```python
votes=df.groupby("Restaurant Name")["Votes"].sum().reset_index(name="Total Votes") 
votes=votes.sort_values(by="Total Votes",ascending=False)
votes
```

```python
ratings.head(5).plot(x="Restaurant Name",y="Avg Rating",kind='bar', color='green')
plt.title("Top 5 Restaurants By Rating")
plt.ylabel("Average Rating")
plt.legend(loc='upper right')
```

```python
votes.head(5).plot(x="Restaurant Name",y="Total Votes",kind='bar', color="indigo")
plt.title("Top 5 Restaurants By Rating")
plt.ylabel("Average Rating")
plt.legend(loc='upper right')
```

## Task:- Analyze the text reviews to identify the most common positive and negative keywords.

```python
txt=df["Rating text"]
txt=txt.value_counts()
txt
```

```python
positive=df.loc[df["Aggregate rating"]> 2.5]
positive
```

```python
positive_keywords=positive["Rating text"].unique()
print(positive_keywords)
```

```python
plt.hist(positive["Rating text"],label=positive_keywords)
plt.title("Most Positive Keywords")
plt.show()
```

```python
negative=df.loc[(df["Aggregate rating"] < 2.5) & (df["Aggregate rating"]>0)]
negative_keyword=negative["Rating text"].unique()
print(negative_keyword)
```

```python
plt.hist(negative["Rating text"])
plt.title("Most Negative Keywords")
plt.show()
```

```python
review_length= df['Rating text'].apply(len)
average_length = review_length.mean().round(2)
print("Average length of reviews:",average_length)
correlation = review_length.corr(df['Aggregate rating']).round(2)
print("Correlation between Review Length and Rating:",correlation)
```

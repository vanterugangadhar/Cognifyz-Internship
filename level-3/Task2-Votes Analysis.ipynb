```python
df=pd.read_csv(r"C:\Users\vante\OneDrive\Desktop\Dataset  cognify og.csv")
df.head()
```

## Task 2: Votes Analysis

```python
df.columns
```

## Task:-Identify the restaurants with the highest and lowest number of votes.

```python
votes=df.groupby('Restaurant Name')['Votes'].sum().sort_values(ascending=False)
highest_voted_restaurants=votes.head(10)
print("The Top 10 Restaurents with highest votes:",highest_voted_restaurants)
```

```python
lowest_voted_restaurants=votes.tail(10)
print("The 10 Restaurents with lowest votes:",lowest_voted_restaurants)
```

## Task:-Analyze if there is a correlation between the number of votes and the rating of a restaurant.

```python
correlation=np.corrcoef(df['Aggregate rating'],df['Votes'])
print(correlation.round(2))
```

```python
sns.heatmap(correlation,annot=True,fmt='.2f')
```

```python
sns.scatterplot(data=df,x='Aggregate rating',y='Votes',hue='Aggregate rating')
sns.regplot(data=df,x='Aggregate rating',y='Votes',scatter=False,color='red')
plt.title('Correlation between Rating and Votes:0.31')
plt.show()
print('As per result correlation is: 0.31.The both variables have positive correlation.') 
```
